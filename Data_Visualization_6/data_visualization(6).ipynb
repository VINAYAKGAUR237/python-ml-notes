# EDA
- EDA: Exploratory Data Analysis
- It is a method of analyzing and summarizing the data before making any assumption or       sometimes performing data cleansing 
- It helps identify patterns, spot anomalies, inspect data quality, test hypotheses, and check for assumptions 
- It is a very imp step in DS/ML pipeline cuz it provides a better understanding of data before performing machine learning modeling
- EDA Techniques: 
    - **Descriptive Statistics** --> measures of central tendency, dispersion, and shape    also correlation
    - **Non-Graphical Analysis** -->
        - data aggregation
        - checking for duplicates
        - checking for outliers

    - **Graphical Analysis** --> building visualization
        - Line graph
        - Pie plot
        - Bar plot
        - Histrogram
        - Heatmaps
        - Boxplots, etc...

- Python Visualization Libraries that we will be using: 
    - 'pandas' visualization
    - 'matplotlib' --> foundation of most python libraries
    - 'seaborn' --> statistical visualization library
    - 'plotly' --> interactive visualization (similar to PowerBI or Tableau)

# Line Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt             #we mostly use pyplot in matplotlib
import seaborn as sns

path = ("/home/vinayakgaur07/Downloads/housing_data.csv")

#generic data
x = np.linspace(0,10,100)                       #numpy array of 100 numbers starting from 0 to 10
y = np.sin(x)

#create a line plot using 'plot()' function:
plt.plot(x, y, color='green', linewidth=4)      #plots x and y  (we can change color & linewidth)

#format the plot area:
plt.xlabel('x-axis')                            #naming x-axis
plt.ylabel('y-axis')                            #naming y-axis
plt.title('Line Plot of a Sine Wave')           #gives titles to plot
```

For various matplotlib themes and styles:
- Matplotlib Styles --> https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
- Matplotlib Colors --> https://matplotlib.org/stable/gallery/color/named_colors.html
  
- For Themes, you only need to run the line once and it will set the theme for all visuals: plt.style.use('ggplot') --> gives grids 

```python
plt.style.use('ggplot')                         #provides grids 

#to chnage dimensions of figure:  ensure this is always at top of program
plt.Figure(figsize=(8,5))           # 8 & 5 are in Inches

#create a line plot using 'plot()' function:
plt.plot(x, y, color='tab:blue', linewidth=4, label='Sine of x')      #plots x & y  (we can change color & linewidth)
#'tab:blue' --> type of color

y2 = np.cos(x)
plt.plot(x, y2, color='orange', linewidth=4, label='Cosine of x', linestyle='--')    #plots x & y2 and also 'linestyle' changes the style of line

# (-) --> this gives Solid linestyle
# (--) --> this gives Dotted linestyle

#format the plot area:
plt.xlabel('x-axis')                            #naming x-axis
plt.ylabel('y-axis')                            #naming y-axis
plt.title('Line Plot of a Sine Wave')   

#to add legend:         legend --> shows what each line represents
plt.legend()            #for this we also have to add 'label' in both plots funcs
plt.legend(loc='upper center')          #this makes legend appear on upper side 
```

**Exercise:**
- Automate using loop, building histograms for 4 numerical entities into 1 canvas

```python
#these are the entities that we can choose from(any 4)
numerical_list
```

```python
numerical_entity=['SalePrice', 'LotArea', 'YearBuilt', 'YrSold']


fig, axes = plt.subplots(nrows=2, ncols=2,  figsize=(12,6))                 #creates canvas

numerical_entity_df=df[numerical_entity]                                    #converts list into dataframe

#to flatten axes position from ([0,0],[0,1],[1,0],[1,1]) to (0,1,2,3):          #helps loop in index
axes = axes.flatten()

'''enumerate --> this will give idx=0, col='SalePrice'
                 then,          idx=1, col='LotArea'
                 then,          idx=2, col='YearBuilt'
                 finally,       idx=3, col='YrSold'
'''

for idx, col in enumerate(numerical_entity):                 #loop 
        sns.histplot(data=numerical_entity_df, x=col, ax=axes[idx])
#                       |                    |              |
#                    dataframe           SalePrice       axes[0]
#                    dataframe           LotArea         axes[1]
#                    dataframe           YearBuilt       axes[2]
#                    dataframe           Yrsold          axes[3]

fig.tight_layout()
```

> if we want one of these to be a boxplot then we do so by using 'if n else' statements:
    and applying condition so that when --> column = LotArea then we get boxplot, otherwise histogram

```python
numerical_entity=['SalePrice', 'LotArea', 'YearBuilt', 'YrSold']
color_list=['orange','red','blue','yellow']


fig, axes = plt.subplots(nrows=2, ncols=2,  figsize=(12,6))                 #creates canvas

numerical_entity_df=df[numerical_entity]                                    #converts list into dataframe

#to flatten axes position from ([0,0],[0,1],[1,0],[1,1]) to (0,1,2,3):
axes = axes.flatten()

'''enumerate --> this will give idx=0, col='SalePrice'
                 then,          idx=1, col='LotArea'
                 then,          idx=2, col='YearBuilt'
                 finally,       idx=3, col='YrSold'
'''

for idx, col in enumerate(numerical_entity):                 #loop 
    if(col == 'LotArea'):
        sns.boxplot(data=numerical_entity_df, x=col, ax=axes[idx], color=color_list[idx])
      
    else:
        sns.histplot(data=numerical_entity_df, x=col, ax=axes[idx], color=color_list[idx])
#                       |                    |              |
#                    dataframe           SalePrice       axes[0]
#                    dataframe           LotArea         axes[1]
#                    dataframe           YearBuilt       axes[2]
#                    dataframe           Yrsold          axes[3]

fig.tight_layout()
```

Visualization above has few issues in aesthetics:
- need to sort elements
- elements are crammed
- more granular x ticks(more values in the grid)
- add title
- add color for each neighborhood(optional)

```python
#adding visual aesthetics:

#changing size:
plt.figure(figsize=(12,6))
sns.countplot(data=df, 
              y='Neighborhood', 
              order= df['Neighborhood'].value_counts().index,     #sorting --> descending order by count 
#                |                                |
#              order                             sorting
              palette='cool') # palette--> gives diff color to each bar 
#values inside palette are listed in matplotlib documentation 

#for x ticks:  spacing of x-axis values --> use 'xticks()'
plt.xticks(range(0,240,10))

#to automate ticks so that we dont have to put exact values: we define end_mark & steps in range 

end_mark = df['Neighborhood'].value_counts().max() + 40     #max() --> gives the maximum value from data that exists

#  '+ 40' --> this adds buffer to max  => so that now our canvas is bigger than max data value from dataset by 40 points

steps = int(end_mark/15)

plt.xticks((range(0,end_mark,steps)))

#title:
plt.title("Analysis Of Neighborhood Frequency")

```

```python
plt.figure(figsize=(12,6))

sns.scatterplot(x=df['SalePrice'], y=df['GrLivArea'], size=df['YearBuilt'], alpha=0.7, hue=df['HeatingQC'], style=df['PavedDrive'])      

#hue= parameter used to categorize data by color-hue adds the color dimension in your plot
#It colors each data point based on the values of a specific column. It lets the plot show a 4th variable using color.
'''Adds color to encode the quality of the heating system, allowing you to see patterns such as:
        a) Houses with excellent heating may cluster at higher prices.
        b) Houses with poor heating may cluster at lower prices.
        c) Any differences in GrLivArea vs SalePrice across heating quality levels.
'''

#style = changes shape to show differences between houses 
'''
circle --> Yes= Pavement Present 
x --> No = Pavement Absent 
square --> Partial = Pavement present partially
'''

plt.ylabel("Above Ground Living Area")                            
```

>Notice how legend appeared automatically

- Houses with excellent heating may cluster at higher prices.
