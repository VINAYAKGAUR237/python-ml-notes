**Exercise Data Visualization**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/home/vinayakgaur07/Downloads/practice_data.csv')
df.head()
```

```python
df['income']=df['income']*10000
df['age']=df['age']*10
```

# Task 1
Build a histogram for every numerical column using subplots and report the columns with most outliers

```python
#identify all numerical columns 
numerical_columns=df.select_dtypes(include='number').columns
numerical_columns
```

```python
plt.style.use('ggplot')         #this gives gridlines 

#create a subplot canvas:
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(15,6))

#now each numerical column as histogram:
#visits
sns.histplot(data=df, x='visits', ax=axes[0,0], color='tab:blue', kde=True)

#nvisits
sns.histplot(data=df, x='nvisits', ax=axes[0,1], color='tab:red', kde=True)

#ovisits
sns.histplot(data=df, x='ovisits', ax=axes[0,2], color='tab:green', kde=True)

#novisits
sns.histplot(data=df, x='novisits', ax=axes[0,3], color='tab:orange', kde=True)

#emergency
sns.histplot(data=df, x='emergency', ax=axes[0,4], color='tab:purple', kde=True)

#hospital
sns.histplot(data=df, x='hospital', ax=axes[1,0], color='tab:cyan', kde=True)

#chronic
sns.histplot(data=df, x='chronic', ax=axes[1,1], color='tab:pink', kde=True)

#age
sns.histplot(data=df, x='age', ax=axes[1,2], color='tab:red', kde=True)

#school
sns.histplot(data=df, x='school', ax=axes[1,3], color='blue', kde=True)

#income
sns.histplot(data=df, x='income', ax=axes[1,4], color='black', kde=True)

fig.tight_layout()
```

# Task 2
Build a Boxplot for all types of visits

```python
#create subplot canvas
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15,6))

#boxplots:
#visits
sns.boxplot(data=df, x='visits', ax=axes[0,0], color='tab:blue')

#nvisits
sns.boxplot(data=df, x='nvisits', ax=axes[0,1], color='tab:red')

#ovisits
sns.boxplot(data=df, x='ovisits', ax=axes[1,0], color='tab:green')

#novisits
sns.boxplot(data=df, x='novisits', ax=axes[1,1], color='tab:orange')

```

# Task 8
Build a correlation heatmap for columns that highly correlate with emergency room visits 

```python
plt.figure(figsize=(12,12))

# sns.heatmap('emergency')

corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix)
```
